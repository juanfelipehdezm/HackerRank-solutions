```python
import pandas as pd
import numpy as np
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date

import warnings
warnings.filterwarnings('ignore')
```

```python
exceldf = pd.read_excel("ejercicio1_b1.xlsx")
del exceldf["\como\lidiar\con\este\campo"]
```

```python
exceldf.dropna(inplace=True)
print(exceldf["id"].isna().sum())
exceldf.drop_duplicates(subset="id", inplace=True)
del exceldf["id"]
exceldf.reset_index(inplace=True)
```

```python
exceldf["Tipo de pedido"].replace({" juguete ":"JUGUETE"}, inplace=True)
exceldf["Tipo de pedido"].replace({" JUGUETE ":"JUGUETE"}, inplace=True)
exceldf["Tipo de pedido"].replace({"Carro ":"CARRO"}, inplace=True)
exceldf["Tipo de pedido"].replace({"ELEC TROD OME STI CO":"ELECTRODOMESTICO"}, inplace=True)
exceldf["Tipo de pedido"].replace({"ELECT RODO MES TI CO":"ELECTRODOMESTICO"}, inplace=True)
exceldf["Tipo de pedido"].replace({"SIN INFO":"SIN_INFO"}, inplace=True)
exceldf["Tipo de pedido"].replace({"mue bles":"MUEBLES"}, inplace=True)
exceldf["Tipo de pedido"].replace({"muebles":"MUEBLES"}, inplace=True)
```

```python
exceldf.rename(columns={"cc_cliente":"CEDULA"}, inplace=True)
del exceldf["index"]
exceldf.reset_index(inplace=True)
```

```python
exceldf
```

```python
grouped_excel_data = exceldf.groupby(["CEDULA","Tipo de pedido"], as_index=False)["numero de pedido"].count()
grouped_excel_data
```

```python
textdf = pd.read_csv("ejercicio1_b2.txt",sep="\t")

textdf["NOMBRE COMPLETO"] = textdf["NOMBRE"] + " " + textdf["APELLIDO"]

textdf["NACIMIENTO"] = pd.to_datetime(textdf["NACIMIENTO"])
base_date = pd.to_datetime("25/11/1988") 

textdf["EDAD"] = abs(textdf["NACIMIENTO"] - base_date)
textdf["EDAD"] = round(textdf["EDAD"]/np.timedelta64(1,"Y"),2)

textdf
```

```python
outputdf = textdf.merge(grouped_excel_data,how = "left", on ="CEDULA")
outputdf
```

```python
outputdf.to_csv("outputdf.txt", sep="\t")
```
